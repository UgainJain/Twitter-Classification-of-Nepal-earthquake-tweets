```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv",header=None)
```

```python
data.head()
```

```python
data = data.fillna(value=" ")
```

```python
data["texts"] = data[2].map(str)+data[3].map(str)+data[4].map(str)+data[5].map(str)+data[6].map(str)+data[7].map(str)+data[8].map(str)
```

```python
data.head()
```

```python
data = data.drop(columns=[2,3,4,5,6,7,8],axis = 1)
```

```python
data["texts"][0]
```

```python
from sklearn.feature_extraction import text
```

```python
vect = text.TfidfVectorizer()
```

```python
x = vect.fit(data["texts"])
```

```python
abc =x.transform(data["texts"])
```

```python
abc
```

```python
vect.vocabulary_
```

```python
stopwords = vect.stop_words_
```

```python
data['texts'] = data['texts'].apply(lambda x: [item for item in str.split(x) if item not in stopwords])
```
